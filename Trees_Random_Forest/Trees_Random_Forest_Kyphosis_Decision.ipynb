# Decision Trees and Random Forests in Python

This is the code which goes over tree methods in Python

```python
# Import the libraries
# Pandas is high-performance, easy-to-use data structures and data analysis tools for the Python programming language
import pandas as pd

# Numpy is the fundamental package for scientific computing with Python
import numpy as np

# Matplotlib is used for graphs
import matplotlib.pyplot as plt

# seaborn Seaborn is a Python visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics
import seaborn as sns

# %matplotlib inline is magic command. This performs the necessary behind-the-scenes setup for IPython to work correctly hand in hand with matplotlib
%matplotlib inline

# Importing train_test_split() for the purpose of spliting the data into test and train sets
from sklearn.model_selection import train_test_split
```

```python
# Reading Ecommerce Kyohosis File 
df = pd.read_csv('kyphosis.csv')
```

```python
# To see first 5 rows of data set
df.head()
```

```python
# To know data type and non null values in a data frame
df.info()
```

```python
# To visualize corelation between different variables
sns.pairplot(df,hue='Kyphosis')
```

```python
# Dividing variables and target variable
X = df.drop('Kyphosis',axis=1)
```

```python
y = df['Kyphosis']
```

```python
# spliting the data into test (30 percent) and train sets (70 percent) with 101 random state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Importing DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Specifying a variable to DecisionTreeClassifier()
dftree = DecisionTreeClassifier()
```

```python
# Creating model using train set
dftree.fit(X_train,y_train)
```

```python
# Predicting test set
predict = dftree.predict(X_test)
```

```python
# Importing classification_report & confusion_matrix check our prediction
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predict))
print(confusion_matrix(y_test,predict))
```

```python
# Importing RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Specifying a variable to RandomForestClassifier()
dfran =RandomForestClassifier(n_estimators=200)
```

```python
# Creating model using train set
dfran.fit(X_train,y_train)

# Predicting test set
predict2 = dfran.predict(X_test)

# Classification_report & confusion_matrix check our prediction
print(classification_report(y_test,predict2))
print(confusion_matrix(y_test,predict2))
```
